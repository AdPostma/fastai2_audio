# The Free Digit Dataset

The [Free Digit Dataset](https://github.com/Jakobovski/free-spoken-digit-dataset) is meant to be an audio version of MNIST, and is great for testing purposes. It has 2000 recordings, 4 speakers saying each digit 50 times. First let's grab the data.

```python
path_digits = untar_data(URLs.DIGITS, extract_func=tar_extract_at_filename)/"recordings"
```

```python
# pull out just the audio files 
files_digits = get_audio_files(path_digits)
```

```python
files_digits[:5]
```

```python
# Open the 6th file and display it
audio_example = AudioTensor.create(files_digits[5])
audio_example.show()
```

## What we pass to our model

While it is completely possible to train on raw audio by passing a 1D tensor of the signal to our model, it is much more easy/common/fast to convert the audio to a spectrogram (a 2D image with frequency data) and train on that.

Here we have an AudioConfig class with different settings for different applications. Since this dataset contains only human voices, we use AudioConfig.Voice() which does things like limit the frequency range to only those present in the human voice.

```python
# get our voice settings from the AudioConfig class
cfg_voice = AudioConfig.Voice()
# generate our spectrogram transformer using the voice config
a2s = AudioToSpec.from_cfg(cfg_voice)
# make a list of transforms that crop the signal to 1000ms and then
# convert it to a spectrogram
crop1s = CropSignal(1000)
tfms = [crop1s, a2s]
```

Let's hear/see the audio signal, and then look at the spectrogram that results from randomly grabbing a 1s chunk of that signal and converting it to a spectrogram

```python
for fn in files_digits[:5]:
    audio = AudioTensor.create(fn)
    audio.show()
    a2s(crop1s(audio)).show()
    print('-'*80)
```

### Training a model using melspectrogram and the DataBlock API

We need to make sure to use...
- an AudioBlock (imported from fastai2_audio.core)
- the appropriate getter function (get_audio_files, also from fastai2_audio.core)
- our audio transforms
- our labeler, which in this case just grabs the first digit of the filename

```python
tfms = [CropSignal(1000), a2s]
```

```python
auds = DataBlock(blocks=(AudioBlock, CategoryBlock),  
                 get_items=get_audio_files, 
                 splitter=RandomSplitter(),
                 item_tfms = tfms,
                 get_y=lambda x: str(x).split('/')[-1][0])
```

```python
dbunch = auds.dataloaders(path_digits, bs=64)
```

```python
dbunch.show_batch()
```

```python
# credit to Kevin Bird and Hiromi Suenaga for these two lines to adjust a CNN model to take 1 channel input
def alter_learner(learn, channels=1):
    learn.model[0][0].in_channels=channels
    learn.model[0][0].weight = torch.nn.parameter.Parameter(learn.model[0][0].weight[:,1,:,:].unsqueeze(1))
```

```python
learn = Learner(dbunch, 
                xresnet18(),
                torch.nn.CrossEntropyLoss(), 
                metrics=[accuracy])
nchannels = dbunch.one_batch()[0].shape[1]
alter_learner(learn, nchannels)
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(5, lr_max=slice(1e-2))
```

```python
learn.fit_one_cycle(5, lr_max=slice(1e-3))
```

95.5% accuracy in 10s, 98% accuracy in 20s, not bad for a small dataset and no data augmentation.

### Using fancier audio trans

To get a good result on this simple dataset, we only need to crop to a single length, and convert to a spectrogram, but I'll add in some transforms that might be useful if you have longer clips with more silence, or if you're overfitting.

Instead of using the default configs for our Spectrogram generator, let's pass in settings manually

```python
#this is the  we still want a decibel scale melspectrogram
DBMelSpec = SpectrogramTransformer(mel=True, to_db=True)
```

```python
a2s = DBMelSpec(n_mels=128, f_max=10000, n_fft=1024, hop_length=128, top_db=100)
```

```python
# MaskTime(size=4), MaskFreq()
tfms = [RemoveSilence(), CropSignal(1000), a2s, MaskTime(size=4), MaskFreq(size=10)]
```

```python
# we could redo the whole auds block, or we can just swap in our new transform pipeline
auds.item_tfms = tfms
# tfms = Pipeline([CropSignal(4000),  a2s, MaskFreq(size=12), MaskTime(size=15), SGRoll()], as_item=True)
dbunch_fancy = auds.dataloaders(path_digits, bs=128)
```

```python
x = dbunch_fancy.show_batch()
```

```python
learn = Learner(dbunch_fancy, 
                xresnet18(),
                torch.nn.CrossEntropyLoss(), 
                metrics=[accuracy])
nchannels = dbunch_fancy.one_batch()[0].shape[1]
alter_learner(learn, nchannels)
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(10, lr_max=slice(5e-3))
```

```python
learn.fit_one_cycle(10, lr_max=slice(5e-3))
```

### Using Mel Frequency Cepstral Coefficient (MFCC)

```python
a2mfcc = AudioToMFCC(n_mfcc=40, melkwargs={"n_fft":2048, "hop_length":256, "n_mels":128})
tfms = [CropSignal(1000), a2mfcc]
# we could redo the whole auds block, or we can just swap in our new transform pipeline
auds.item_tfms = tfms
# tfms = Pipeline([CropSignal(4000),  a2s, MaskFreq(size=12), MaskTime(size=15), SGRoll()], as_item=True)
dbunch_mfcc = auds.dataloaders(path_digits, bs=128)
```

Note: These images are actually only 40px tall, usually 20 or 40 are the default for n_mfcc. 

```python
dbunch_mfcc.show_batch()
```

```python
learn = Learner(dbunch_mfcc, 
                xresnet18(),
                torch.nn.CrossEntropyLoss(), 
                metrics=[accuracy])
nchannels = dbunch_mfcc.one_batch()[0].shape[1]
alter_learner(learn, nchannels)
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(5, lr_max=slice(1e-2))
```

```python
learn.fit_one_cycle(5, lr_max=slice(1e-3))
```
